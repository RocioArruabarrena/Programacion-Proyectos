```python
# Ejercicio 1
# Ordenación de Productos
# Dado un diccionario donde las claves son nombres de productos y los valores son tuplas con (precio, cantidad en stock), 
# escribe un programa que ordene los productos de mayor a menor precio y devuelva una lista ordenada de tuplas con (nombre, precio).


productos = {
    "Laptop": (1200, 5),
    "Smartphone": (800, 10),
    "Tablet": (500, 7),
    "Monitor": (300, 4),
    "Teclado": (50, 20)
}

prod_ordenados = sorted (productos.items(), key= lambda item: item [1][0], reverse=True) 
#productos.items devuelve una tupla, key=lambda item: ordena los elementos y extrae el precio de cada producto, item[1] tiene el precio y el stock, 
# sorted ordena de menor a mayor y reverse=true ordena de mayor a menor lo ordenado por sorted

resultado = [(nombre, precio[0])for nombre, precio in prod_ordenados]
#for nombre, precio recorre la lista y precio [0] extrae el primer valor d ela tupla creada
print(resultado)

```

```python
# Ejercicio 2
# Agrupación de Datos
# Escribe una función que reciba una lista de tuplas con (nombre, edad, ciudad) y devuelva un diccionario donde las claves 
# sean las ciudades y los valores sean listas con los nombres de las personas que viven en esa ciudad.

def ordenar_por_ciudades(personas):
    ordenado = {} #crear el diccionario
        
    for nombre, edad, ciudad in personas: #recorre la lista de tupolas 
        if ciudad not in ordenado: #si la ciudad no esta en el diccionario la agregamos
          ordenado[ciudad] = []
        ordenado[ciudad].append(nombre) #append agrega el nombre a la ciudad que correspinda
    return ordenado #devuelve el diccionario
            
personas = [
    ("Juan", 25, "Madrid"),
    ("Ana", 30, "Barcelona"),
    ("Luis", 22, "Madrid"),
    ("Elena", 27, "Sevilla"),

    ("Carlos", 35, "Barcelona")
]        
            
print(ordenar_por_ciudades(personas))


```

```python
# Ejercicio 3
# Cálculo de Promedios
# Crea una función que reciba un diccionario donde las claves sean nombres de estudiantes y los valores sean listas de calificaciones. 
# La función debe devolver un nuevo diccionario con los promedios de cada estudiante redondeados a dos decimales.

def promedios_estudiante(calificaciones):
    promedios = {}
    
    for estudiante, notas in calificaciones.items():
         promedio = sum(notas) / len(notas) #calculamos el promedio
         promedios[estudiante] = round(promedio, 2) #promediamos y redondiamos a dos decimales 
    return promedios 
      

calificaciones = {
    "Maria": [85, 90, 78],
    "Jose": [70, 88, 92],
    "Lucia": [95, 85, 87],
    "Pablo": [60, 75, 80]
}

print(promedios_estudiante(calificaciones)) 
```

```python
# Ejercicio 4
# Frecuencia de Palabras
# Dado un texto ingresado por el usuario, escribe un programa que cuente la frecuencia de cada palabra y almacene los resultados en un diccionario. 
# La salida debe mostrar las palabras ordenadas de mayor a menor frecuencia.

def conteo_palabras(texto):
    palabras = texto.split () #el split divide el texto en una lista d epalabras
    frecuencia = {} 
    
    for palabra in palabras:  
        frecuencia[palabra] = frecuencia.get(palabra, 0) + 1 #vamos agregando a el contador de la palabra
    
    return dict(sorted(frecuencia.items(), key=lambda item: item[1], reverse=True)) #ordenamos con sorted de menor a mayor y con reverse=true de mayor a menor frecuencia el diccionario

texto = "python es un lenguaje de programación python es poderoso y versátil"


print(conteo_palabras(texto))
```

```python
# Ejercicio 6
# Análisis de Ventas
# Dado un diccionario donde las claves son nombres de vendedores y los valores son listas de ventas realizadas en el mes, 
# escribe una función que determine qué vendedor tuvo la mayor suma total de ventas.

def mejor_vendedor(ventas):
    total_ventas = {vendedor:sum(ventas_list) for vendedor, ventas_list in ventas.items()} #ventas de cada empleado sumadas 
    mejor = max(total_ventas,key= total_ventas.get) #con el max buscamos en la suma anterior el vendedor que tiene mas ventas
    
    return mejor, total_ventas[mejor] # con el return devolvemos quien es el mejor empleado junto a su total de ventas
    

ventas = {
    "Pedro": [200, 150, 400],
    "Lucia": [300, 500, 700],
    "Ana": [100, 300, 200]
}

print(mejor_vendedor(ventas))#imprimimos el resultado
```

```python
# Ejercicio 7
# Conversión de Listas a Diccionario
# Crea una función que reciba dos listas: una con nombres de personas y otra con sus edades. 
# La función debe devolver un diccionario donde los nombres sean las claves y las edades sean los valores.

def lista_a_dict (nombres,edades): #definimos el diccionario para ambas listas
    return dict (zip(nombres,edades)) #el return devuelve el diccionario creado y con la funcion zip juntamos ambas listas

nombres = ["Carlos", "Ana", "Pedro", "Marta"]
edades = [28, 34, 25, 40]

resultado = lista_a_dict (nombres,edades) #guardamos el diccionario dentro de la funcion de resultado 

print(resultado) #imprimimos el resultado
```

```python
# Ejercicio 8
# Filtrado de Datos
# Escribe una función que reciba un diccionario con nombres de ciudades y temperaturas en grados Celsius. 
# La función debe devolver un nuevo diccionario con solo las ciudades donde la temperatura sea superior a 30°C.

def filtrar_datos_ciudad (temperaturas): #definir 
    
    return {ciudad: temp for ciudad, temp in temperaturas. items() if temp > 30}  #el .item devuelve una lista de tuplas y lo demas devuelve el diccionario por comprension, si la temp es mayor a 30 grados la imprime


temperaturas = {
    "Madrid": 32,
    "Barcelona": 28,
    "Sevilla": 35,
    "Bilbao": 22
}

resultado = filtrar_datos_ciudad(temperaturas) #guardamos el diccionario en resultado 

print(resultado)
```

```python
# Ejercicio 10
# Análisis de Encuestas
# Se tiene una lista de tuplas donde cada tupla representa la respuesta de un usuario a una encuesta (nombre, edad, respuesta).  
# Escribe una función que devuelva un diccionario donde las claves sean las respuestas y los valores sean listas con los nombres de las 
# personas que eligieron esa respuesta.


def analisis_de_encuentas(respuestas):
    valores = {} #crear diccionario
    for nombre,edad, respuesta in respuestas: #por los valores en la lista 
        if respuesta not in valores: #si la respuesta no esta en el diccionario de valores, 
           valores [respuesta] = [] #se crea una nueva entrada
           valores [respuesta].append(nombre) #se agrega el nombre del participante a la lista que corresponda con la funcion .append
    return valores #se devuelve el diccionario 
     

respuestas = [
    ("Carlos", 25, "Opción A"),
    ("Ana", 30, "Opción B"),
    ("Pedro", 22, "Opción A"),
    ("Marta", 27, "Opción C"),
    ("Lucia", 35, "Opción B")
]


resultado = analisis_de_encuentas (respuestas) #guardar el diccionario en la funcion resultado

print (resultado)#imprimir resultado


#corregir porque no salen todos los nombres correspondientes a la respuesta en los valores del dict
```
